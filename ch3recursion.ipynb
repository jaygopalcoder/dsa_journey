```python
def factorial(n:int)->int:
    if n<0:
        return -1
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n*factorial(n-1)

factorial(-12334)
```

```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None

class MyLinkedList:

    def __init__(self):
        self.head = None

    def get(self, index: int) -> int:
        self.index = index
        if self.index<0 or self.head==None:
            return -1
        count = -1
        current = self.head
        while current is not None:
            count+=1
            if count == self.index:
                return current.val
            current = current.next

    def insert_at_beg(self, data):
        temp = Node(data)
        if not self.head:
            self.head = temp
        else:
            temp.next = self.head
            self.head = temp

mll = MyLinkedList()
mll.insert_at_beg(34)
mll.insert_at_beg(7)
mll.insert_at_beg(78)
mll.get(2)
```
